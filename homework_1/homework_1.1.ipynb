```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
from math import pi

pi, np.pi
```

```python
def psi_sq(x):
    psi = np.exp(- x ** 2. / 2.) / np.pi**(1/4)
    return abs(psi) ** 2.
```

```python
N_steps = 100000
delta = 0.1
x = 0.0

xs = []
for i in range(N_steps):
    x_new = x + random.uniform(-delta, delta)
    if random.uniform(0.0, 1.0) < (psi_sq(x_new)) / (psi_sq(x)) :
        #(prob(x_new, n)/prob(x, n)):
        x = x_new
        xs.append(x)
```

```python
pylab.hist(xs, 500, density = 'True', label='Markov-chain sampling') #histogram of the sample
x = [a / 10.0 for a in range(-30, 30)]
y = [psi_sq(a) for a in x]
pylab.plot(x, y, c='red', linewidth=2.0, label='Exact quantum')
pylab.title('Position distribution at $T=0$', fontsize = 13)
pylab.xlabel('$x$', fontsize = 15)
pylab.ylabel('$\pi(x)=|\psi_0(x)|^2$', fontsize = 15)
pylab.legend()
```
